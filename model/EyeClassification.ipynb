```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

def get_classification_model(input_shape, num_classes):
    model = Sequential()

    # Convolutional layers
    model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=input_shape))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    
    model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    
    model.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))

    model.add(Flatten())
    
    # Fully connected layers
    model.add(Dense(512, activation='relu'))
    model.add(Dropout(0.5))
    
    model.add(Dense(1, activation='sigmoid'))
    return model

# Define input shape and number of classes
input_shape = (256, 256, 3)  # Adjust the dimensions based on your image size
num_classes = 2  # Two classes for binary classification

# Create the classification model
classification_model = get_classification_model(input_shape, num_classes)

# Compile the model
classification_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Print the summary of the model architecture
classification_model.summary()
```

```python
import pandas as pd
import numpy as np
from PIL import Image
import os

def load_data(path):
    data = pd.read_csv(r"C:\Users\msria\Downloads\own_data_1.csv")
    new_data = data.astype('str')

    x = []
    y = []
    total = len(data)
    for i in range(total):
        id_name = new_data.iloc[i]['ID']
        im_p_jpg = id_name + '.jpg'
        im_p_png = id_name + '.png'
        
        path_jpg = os.path.join(path, im_p_jpg)
        path_png = os.path.join(path, im_p_png)
        
        if os.path.exists(path_jpg):
            im_p = im_p_jpg
        elif os.path.exists(path_png):
            im_p = im_p_png
        else:
            print(f"Image not found for ID: {id_name}")
            continue
        
        im = Image.open(os.path.join(path, im_p))
        im = im.resize((256, 256))
        im = np.array(im).astype('float16')
        im = im / 255 - 0.5
        print(im.shape)
        x.append(im)
        y.append(data.iloc[i]['Class'])

    return x, y
```

```python
import numpy as np
import cv2
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Simulated eye detection data (example)
# Each row: [x_min, y_min, x_max, y_max, class]
# Assume normalized coordinates (0 to 1)
eye_annotations = np.array([
    [0.2, 0.3, 0.3, 0.4, 1],
    [0.6, 0.4, 0.7, 0.5, 1]
])

# Load and preprocess image
image = cv2.imread(r"C:\Users\msria\Downloads\sriRagu.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = image / 255.0

# Define input shape and number of classes
input_shape = (256, 256, 3)
num_classes = 2  # 1 for eyes, 1 for background

# Create the detection model
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=input_shape))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(num_classes + 4))  # 4 for bounding box coordinates

# Define custom loss function (example, adapt as needed)
def detection_loss(y_true, y_pred):
    # Split predictions into class and bounding box predictions
    y_class_true = y_true[:, :num_classes]
    y_class_pred = y_pred[:, :num_classes]
    y_bbox_true = y_true[:, num_classes:]
    y_bbox_pred = y_pred[:, num_classes:]

    # Classification loss
    class_loss = tf.keras.losses.categorical_crossentropy(y_class_true, y_class_pred)

    # Bounding box regression loss (example)
    bbox_loss = tf.reduce_mean(tf.square(y_bbox_true - y_bbox_pred))

    # Combine losses (adjust weighting as needed)
    total_loss = class_loss + 10 * bbox_loss
    return total_loss

# Compile the model
model.compile(optimizer='adam', loss=detection_loss)

# Prepare annotations for training
y_true = np.zeros((1, num_classes + 4))  # One example with 1 class + 4 bounding box values
for annotation in eye_annotations:
    x_min, y_min, x_max, y_max, label = annotation
    bbox_center_x = (x_min + x_max) / 2
    bbox_center_y = (y_min + y_max) / 2
    bbox_width = x_max - x_min
    bbox_height = y_max - y_min

    y_true[0, int(label)] = 1  # Set class label
    y_true[0, num_classes:num_classes + 4] = [bbox_center_x, bbox_center_y, bbox_width, bbox_height]

# Train the model
model.fit(image.reshape((1,) + input_shape), y_true, epochs=10)

# Inference (example)
predictions = model.predict(image.reshape((1,) + input_shape))
class_predictions = predictions[0, :num_classes]
bbox_predictions = predictions[0, num_classes:]

print("Class Predictions:", class_predictions)
print("Bounding Box Predictions:", bbox_predictions)
```
